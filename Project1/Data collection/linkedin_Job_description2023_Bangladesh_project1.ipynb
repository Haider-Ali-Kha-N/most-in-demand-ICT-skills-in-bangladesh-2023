```python
#importing selenium and other classes & dependencies
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

#creating object to connect with chrome driver
service_obj=Service(r"C:/chromedriver_win32/chromedriver.exe")
driver=webdriver.Chrome(service=service_obj)
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
import time
driver.implicitly_wait(3)

#loging into linkedin
email="******@gmail.com"
password="****"
driver.get("https://www.linkedin.com/home")
driver.find_element(By.XPATH,"//input[@type='text']").send_keys(email)
driver.find_element(By.XPATH,"//input[@type='password']").send_keys(password)
driver.find_element(By.XPATH,"//button[@class='sign-in-form__submit-button']").click()
```

```python
#going to the page of jobs in BD
driver.get("https://www.linkedin.com/jobs/search/?currentJobId=3450317869&geoId=106215326&location=Bangladesh&refresh=true")
```

```python
#importing pandas and creating lists for columns
import pandas as pd
df=pd.DataFrame()
ljob_title=[]
lname_location=[]
lexperience_level=[]
ljob_description=[]
```

```python
#looping through the pages
for j in range(25,41):
    page_numbers=driver.find_element(By.XPATH,f"//button[@aria-label='Page {j}']" )
    page_numbers.click()
    time.sleep(3)

    jobs_block= driver.find_elements(By.XPATH,"//ul[@class='scaffold-layout__list-container']/li")
    
    #scrolling to the bottom of each page so that all the text load
    for job in jobs_block:
        # executes JavaScript to scroll the div into view
        driver.execute_script("arguments[0].scrollIntoView();", job)
        time.sleep(.1)

#looping and scraping job data from page j
    job_links=driver.find_elements(By.XPATH,"//ul[@class='scaffold-layout__list-container']/li/div/div/div[1]/div[2]/div[1]/a")
    for link in job_links:
        link.click()
        #waiting to load
        time.sleep(3)
        job_title=driver.find_element(By.XPATH,"//h2[@class='t-24 t-bold jobs-unified-top-card__job-title']").text
        name_location=driver.find_element(By.XPATH,"//span[@class='jobs-unified-top-card__subtitle-primary-grouping t-black']").text
        experience_level=driver.find_element(By.XPATH,"//div[@class='mt5 mb2']/ul/li[1]").text
        job_description=driver.find_element(By.XPATH,"//div[@id='job-details']").text

        ljob_title.append(job_title)
        lname_location.append(name_location)
        lexperience_level.append(experience_level)
        ljob_description.append(job_description)
```

```python
#creating dataframe
data={'job_title':ljob_title,'name_location':lname_location,'experience_level':lexperience_level,'job_description':ljob_description}
df=pd.DataFrame(data)
print(df)
```
