```python
#importing selenium and other classes & dependencies
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

#creating object to connect with chrome driver
service_obj=Service(r"C:/chromedriver_win32/chromedriver.exe")
driver=webdriver.Chrome(service=service_obj)
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
import time
```

```python
#going to the page of jobs in BD
driver.get("https://jobs.bdjobs.com/jobsearch.asp?fcatId=8")
```

```python
#creating a list for storing collected links of jobs
job_links_list=[]
```

```python
#loop through the pages
for i in range(1,27):
    page=driver.find_element(By.XPATH,f"//div[@id='topPagging']//a[contains(text(),{i})]")
    page.click()
    time.sleep(3)
    job_links=driver.find_elements(By.XPATH,"//a[@onclick='clickJObTitle()']")
    
    #collecting links using 'href'
    for link in job_links:
        job_links_list.append(link.get_attribute('href'))
```

```python
#importing pandas
import pandas as pd
```

```python
#checking
print(job_links_list)
```

```python
#creating lists for columns
ltitle=[]
lcompany=[]
ljob_info1=[]
lvacancy=[]
ljob_context=[]
ljob_res=[]
lemp_status=[]
ledu_req=[]
lexp_req=[]
ladd_req=[]
llocation=[]
lsalary=[]
lother_bnf=[]
lcomp_details=[]
```

```python
#collecting the necessary data

for link in job_links_list:

    driver.get(link)
    time.sleep(4)

    try:
        title=driver.find_element(By.XPATH,"//div[@class='col-sm-9']/h2").text
        ltitle.append(title)
    except:
        title="n/a"
        ltitle.append(title)

    try:
        compnay=driver.find_element(By.XPATH,"//div[@class='col-sm-9']/h2").text
        lcompany.append(compnay)
    except:
        company="n/a"
        lcompany.append(compnay)
    try:    
        job_info1=driver.find_element(By.XPATH,"//div[@class='Job_keyPoints']").text
        ljob_info1.append(job_info1)
    except:
        job_info1="n/a"
        ljob_info1.append(job_info1)
    try:
        vacancy=driver.find_element(By.XPATH,"//div[@class='vac']").text
        lvacancy.append(vacancy)
    except:
        vacancy="n/a"
        lvacancy.append(vacancy)
    try:
        job_context=driver.find_element(By.XPATH,"//div[@class='job_des'][1]").text
        ljob_context.append(job_context)
    except:
        job_context="n/a"
        ljob_context.append(job_context)
    try:   
        job_res=driver.find_element(By.XPATH,"//div[@class='job_des'][2]").text
        ljob_res.append(job_res)
    except:
        job_res="n/a"
        ljob_res.append(job_res)
    try:
        emp_status=driver.find_element(By.XPATH,"//div[@class='job_nat']").text
        lemp_status.append(emp_status)
    except:
        emp_status="n/a"
        lemp_status.append(emp_status)
    try:
        edu_req=driver.find_element(By.XPATH,"//div[@class='edu_req'][1]").text
        ledu_req.append(edu_req)
    except:
        edu_req="n/a"
        ledu_req.append(edu_req)
    try:
        exp_req=driver.find_element(By.XPATH,"//div[@class='edu_req'][2]").text
        lexp_req.append(exp_req)
    except:
        exp_req="n/a"
        lexp_req.append(exp_req)
    try:
        add_req=driver.find_element(By.XPATH,"//div[@class='job_req']").text
        ladd_req.append(add_req)
    except:
        add_req="n/a"
        ladd_req.append(add_req)
    try:
        location=driver.find_element(By.XPATH,"//div[contains(@class,'job_loc')]").text
        llocation.append(location)
    except:
        location="n/a"
        llocation.append(location)
    try:
        salary=driver.find_element(By.XPATH,"//div[@class='salary_range']").text
        lsalary.append(salary)
    except:
        salary="n/a"
        lsalary.append(salary)
    try:
        other_bnf=driver.find_element(By.XPATH,"//div[@class='oth_ben']").text
        lother_bnf.append(other_bnf)
    except:
        other_bnf="n/a"
        lother_bnf.append(other_bnf)
    try:
        comp_details=driver.find_element(By.XPATH,"//div[@class='information']").text
        lcomp_details.append(comp_details)
    except:
        comp_details="n/a"
        lcomp_details.append(comp_details)
```

```python
#creating dataframe
data={"title":ltitle,"company":lcompany,"job_info1":ljob_info1,"vacancy":lvacancy,"job_context":ljob_context,"job_res":ljob_res,"emp_status":lemp_status,"edu_req":ledu_req,"exp_req":lexp_req,"add_req":ladd_req,"location":llocation,"salary":lsalary,"other_bnf":lother_bnf,"comp_details":lcomp_details,"job_links":job_links_list}
df=pd.DataFrame(data)
```
